```python
import pandas as pd
import numpy as np
import glob
%matplotlib inline
```

```python
onsets_dir = 'C:/Users/jgray7700/Usersjgray7700/onsets_fingerfootlips.csv'

func_data = glob.glob('C:/Users/jgray7700/Usersjgray7700/fmriprep/sub-*/ses-test/func/sub-*_ses-test_task-fingerfootlips_bold_space-MNI152NLin2009cAsym_preproc.nii.gz')
confounds = glob.glob('C:/Users/jgray7700/Usersjgray7700/fmriprep/sub-*/ses-test/func/sub-*_ses-test_task-fingerfootlips_bold_confounds.tsv')

print(func_data)
print(confounds)

nr_subj=len(func_data)
```

```python
motion_params = []
for n in range(nr_subj):
    temp_df = pd.read_csv(confounds[n], delimiter='\t')
    sub_motion_params = pd.DataFrame(temp_df, columns=['X', 'Y', 'Z', 'RotX', 'RotY', 'RotZ'])
    motion_params.append(sub_motion_params)
```

Smoothing the data
------------------------------------

```python
from nilearn import image
func_data_sm = []
for n in range (nr_subj):
    func_data_sm.append(image.smooth_img(func_data[n], fwhm=8))

from nilearn.plotting import plot_epi
mean_img = image.mean_img(func_data_sm[0])
plot_epi(mean_img, title='Smoothed mean EPI')

mean_img1 = image.mean_img(func_data[0])
plot_epi(mean_img1, title='unsmoothed mean EPI')
```

Specifying the experimental paradigm
------------------------------------

We must provide a description of the experiment, that is, define the
timing of the auditory stimulation and rest periods. According to
the documentation of the dataset, there were 16 42s blocks --- in
which 6 scans were acquired --- alternating between rest and
auditory stimulation, starting with rest. We use standard python
functions to create a pandas.DataFrame object that specifies the
timings:

```python
tr = 2.5
n_scans = 178

events = pd.read_csv(onsets_dir)
events['onset'] -= 10
events
```

```python
from nilearn import image 
func_data_sm = [] # initialize a list to put smoothed data
for n in range(nr_subj):
    func_data_sm.append(image.smooth_img(func_data[n], fwhm=8).slicer[..., 4:])
```

```python
from nistats.first_level_model import FirstLevelModel

events_test = [events] * len(func_data_sm)

fmri_glm = FirstLevelModel(tr, drift_model = None)
fmri_glm = fmri_glm.fit(func_data_sm, events_test, confounds = motion_params)
```

```python
from nistats.reporting import plot_design_matrix
design_matrix = fmri_glm.design_matrices_[0]
plot_design_matrix(design_matrix)
```

```python
contrast_matrix = np.eye(design_matrix.shape[1])
contrasts = dict([(column, contrast_matrix[i])
                  for i, column in enumerate(design_matrix.columns)])
contrast_matrix
```

```python
lips_zmap = fmri_glm.compute_contrast('Lips', output_type='z_score')
lips_effmap = fmri_glm.compute_contrast('Lips', output_type='effect_size')

foot_zmap = fmri_glm.compute_contrast('Foot', output_type='z_score')
foot_effmap = fmri_glm.compute_contrast('Foot', output_type='effect_size')

finger_zmap = fmri_glm.compute_contrast('Finger', output_type='z_score')
finger_effmap = fmri_glm.compute_contrast('Finger', output_type='effect_size')
```

```python
lips_minus_foot =  contrasts['Lips'] - contrasts['Foot']

lips_min_foot_eff_map = fmri_glm.compute_contrast(lips_minus_foot,
                                    output_type='effect_size')

lips_min_foot_z_map = fmri_glm.compute_contrast(lips_minus_foot,
                                  output_type='z_score')

finger_minus_foot =  contrasts['Finger'] - contrasts['Foot']

finger_min_foot_effmap = fmri_glm.compute_contrast(finger_minus_foot,
                                    output_type='effect_size')

finger_min_foot_z_map = fmri_glm.compute_contrast(finger_minus_foot,
                                  output_type='z_score')
```

# from nilearn.plotting import plot_stat_map, plot_anat, plot_img

```python
from nilearn.plotting import plot_stat_map

plot_stat_map(lips_zmap, threshold=3.0,
              display_mode='z', cut_coords=9, black_bg=True,
              title='Lips (Z>3)')

plot_stat_map(finger_zmap, threshold=3.0,
              display_mode='z', cut_coords=9, black_bg=True,
              title='Finger (Z>3)')

plot_stat_map(foot_zmap, threshold=3.0,
              display_mode='z', cut_coords=9, black_bg=True,
              title='Foot (Z>3)')

plot_stat_map(lips_min_foot_z_map, threshold=3.0,
              display_mode='z', cut_coords=9, black_bg=True,
              title='Lips minus Foot (Z>3)')

plot_stat_map(finger_min_foot_z_map, threshold=3.0,
              display_mode='z', cut_coords=9, black_bg=True,
              title='Finger minus Foot (Z>3)')
```
